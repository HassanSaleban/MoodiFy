```python
# Définir la taille de la figure pour une meilleure lisibilité
plt.figure(figsize=(10, 2.5))

# Affichage de la boîte à moustaches pour la variable "loudness"
# - showmeans=True : affiche la moyenne sur le graphique
# - width=0.3 : ajuste la largeur de la boîte pour l'esthétique
# - boxprops, medianprops, meanprops, whiskerprops, capprops, flierprops : personnalisent les couleurs et styles pour bien distinguer chaque métrique
sns.boxplot(
    data=df_spotify,
    x="loudness",
    width=0.3,
    showmeans=True,
    meanprops={
        "marker": "o",
        "markerfacecolor": "crimson",
        "markeredgecolor": "black",
        "markersize": 8,
        "linestyle": "--"
    },
    boxprops={"facecolor": "lightblue", "edgecolor": "navy"},
    medianprops={"color": "orange", "linewidth": 3},
    whiskerprops={"color": "black", "linewidth": 2},
    capprops={"color": "black", "linewidth": 2},
    flierprops={
        "marker": "d",
        "markerfacecolor": "red",
        "markeredgecolor": "black",
        "markersize": 6,
        "alpha": 0.7
    }
)

# Calcul des valeurs statistiques pour annotation
q1 = df_spotify["loudness"].quantile(0.25)
median = df_spotify["loudness"].median()
q3 = df_spotify["loudness"].quantile(0.75)
mean = df_spotify["loudness"].mean()
iqr = q3 - q1
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

# Ajout d'annotations pour chaque métrique importante
plt.axvline(mean, color='crimson', linestyle='--', linewidth=2, label='Moyenne')
plt.axvline(median, color='orange', linestyle='-', linewidth=2, label='Médiane')
plt.axvline(q1, color='green', linestyle=':', linewidth=2, label='Q1 (25%)')
plt.axvline(q3, color='blue', linestyle=':', linewidth=2, label='Q3 (75%)')
plt.axvline(lower_whisker, color='purple', linestyle='-.', linewidth=1.5, label='Whisker bas')
plt.axvline(upper_whisker, color='purple', linestyle='-.', linewidth=1.5, label='Whisker haut')

# Ajout de la légende et du titre
plt.legend(loc='upper right')
plt.title("Boîte à moustaches de la variable 'loudness' avec métriques annotées", fontsize=14, fontweight='bold')
plt.xlabel("Loudness (dB)")
plt.ylabel("")

# Affichage du graphique final
plt.tight_layout()
plt.show()
```

```python
# Trouver l'index de la colonne "key"
key_index = df_spotify.columns.get_loc("key")  # get_loc retourne l'index de la colonne "key"

# Supprimer la colonne "unaltered_key" si elle existe déjà pour éviter une erreur lors de l'insertion
if "unaltered_key" in df_spotify.columns:
    df_spotify = df_spotify.drop("unaltered_key", axis=1)

# Insérer la colonne "unaltered_key" juste après "key"
df_spotify.insert(
    key_index + 1,                        # Position d'insertion (après "key")
    "unaltered_key",                      # Nom de la nouvelle colonne
    df_spotify["key"].copy()              # Valeurs à insérer (copie de la colonne "key")
)

df_spotify
```

```python
# Suppression des dièses dans la colonne "unaltered_key" avec apply() et lambda + regex

df_spotify["unaltered_key"] = df_spotify["unaltered_key"].apply(lambda x: re.sub(r'#', '', x))
df_spotify
```

```python
# Importation des bibliothèques nécessaires
import matplotlib.pyplot as plt
import seaborn as sns

# Définition du style visuel pour le graphique
sns.set_style("whitegrid")

# Création de l'histogramme de la variable 'speechiness'
# - bins=30 : nombre de classes pour une meilleure résolution
# - color='mediumseagreen' : couleur des barres de l'histogramme
# - stat='density' : normalisation de l'axe Y pour afficher la densité
# - edgecolor='black' : contour noir autour des barres pour la lisibilité
# - alpha=0.8 : légère transparence des barres
sns.histplot(
    data=df_spotify,
    x="speechiness",
    bins=30,
    color='mediumseagreen',
    stat='density',
    edgecolor='black',
    alpha=0.8
)

# Ajout d'une courbe de densité (KDE) par-dessus l'histogramme
# - color='crimson' : couleur rouge pour la courbe de densité
# - linewidth=2 : épaisseur de la courbe pour la rendre visible
sns.kdeplot(
    data=df_spotify,
    x="speechiness",
    color='crimson',
    linewidth=2
)

# Ajout du titre et des labels des axes
plt.title('Distribution de la colonne speechiness', fontsize=15, fontweight='bold')
plt.xlabel('Speechiness', fontsize=12)
plt.ylabel('Densité', fontsize=12)

# Limitation de l'axe X à l'intervalle [0, 1] pour cadrer la variable
plt.xlim(0, 1)

# Affichage du graphique
plt.show()
```

```python
# Affichage d'une boite à moustache pour la variable "speechiness"
# Mettre en évidence la présence d'outliers

sns.boxplot (data=df_spotify, x="speechiness")
plt.show()
```

```python
# Définir la taille de la figure pour une meilleure lisibilité
plt.figure(figsize=(10, 2.5))

# Affichage de la boîte à moustaches pour la variable "speechiness"
# - showmeans=True : affiche la moyenne sur le graphique
# - width=0.3 : ajuste la largeur de la boîte pour l'esthétique
# - boxprops, medianprops, meanprops, whiskerprops, capprops, flierprops : personnalisent les couleurs et styles pour bien distinguer chaque métrique
sns.boxplot(
    data=df_spotify,
    x="speechiness",
    width=0.3,
    showmeans=True,
    meanprops={
        "marker": "o",
        "markerfacecolor": "crimson",
        "markeredgecolor": "black",
        "markersize": 8,
        "linestyle": "--"
    },
    boxprops={"facecolor": "lightblue", "edgecolor": "navy"},
    medianprops={"color": "orange", "linewidth": 3},
    whiskerprops={"color": "black", "linewidth": 2},
    capprops={"color": "black", "linewidth": 2},
    flierprops={
        "marker": "d",
        "markerfacecolor": "red",
        "markeredgecolor": "black",
        "markersize": 6,
        "alpha": 0.7
    }
)

# Calcul des valeurs statistiques pour annotation
q1 = df_spotify["speechiness"].quantile(0.25)
median = df_spotify["speechiness"].median()
q3 = df_spotify["speechiness"].quantile(0.75)
mean = df_spotify["speechiness"].mean()
iqr = q3 - q1
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

# Ajout d'annotations pour chaque métrique importante
plt.axvline(mean, color='crimson', linestyle='--', linewidth=2, label='Moyenne')
plt.axvline(median, color='orange', linestyle='-', linewidth=2, label='Médiane')
plt.axvline(q1, color='green', linestyle=':', linewidth=2, label='Q1 (25%)')
plt.axvline(q3, color='blue', linestyle=':', linewidth=2, label='Q3 (75%)')
plt.axvline(lower_whisker, color='purple', linestyle='-.', linewidth=1.5, label='Whisker bas')
plt.axvline(upper_whisker, color='purple', linestyle='-.', linewidth=1.5, label='Whisker haut')

# Ajout de la légende et du titre
plt.legend(loc='upper right')
plt.title("Boîte à moustaches de la variable 'speechiness' avec métriques annotées", fontsize=14, fontweight='bold')
plt.xlabel("Speechiness")
plt.ylabel("")

# Affichage du graphique final
plt.tight_layout()
plt.show()
```
